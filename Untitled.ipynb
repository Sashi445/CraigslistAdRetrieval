```python
import requests
from bs4 import BeautifulSoup
```

```python
def page_info(page_url):
    
    page_data = requests.get(page_url)
    
    soup = BeautifulSoup(page_data.content, "html.parser")

    title = soup.find(class_="postingtitle")

    posting_body = soup.find("section", {"id" : "postingbody"})

    print(title.text.strip())
    print(posting_body.text.strip())
    
    print("\n\n\nEND OF POSTING\n\n\n")


```

```python
def field_search(page_url):
    
    page = requests.get(page_url)

    soup = BeautifulSoup(page.content, "html.parser")

    all_links = soup.find_all(class_="result-row")

    for link in all_links[:3]:
        anchor = link.find('a')['href']
        page_info(anchor)


```

```python
FIELDS = ['ccc', 'eee', 'sss', 'ggg', 'hhh', 'jjj', 'rrr', 'bbb']

for field in FIELDS :
    
    page_url = f"https://bangalore.craigslist.org/search/{field}?"
    
    field_search(page_url)
    


```
